```python
import glob
import numpy as np
from imageio import imread
import matplotlib.pyplot as plt

images = []
for filename in glob.glob('images/*.jpeg'):
    image = imread(filename)
    images.append(image)
    
images = np.array(images) / 255
    
print(f"Loaded {len(images)} images")
```

```python
# let's start with a very very simple experiment:
# let's try to create a decoded for a single image using a single value
```

```python
plt.imshow(images[2])
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Reshape, Conv2D, UpSampling2D
```

```python
def Simple_decoder():
    model = Sequential()
    model.add(Dense(64*64*64, input_shape=(1,))) # we'll pass a single value as input
    model.add(Activation('relu'))
    model.add(Reshape((64, 64, 64)))
    model.add(Conv2D(32, 3, padding='same'))
    model.add(Activation('relu'))
    model.add(UpSampling2D(2))
    model.add(Conv2D(16, 3, padding='same'))
    model.add(Activation('relu'))
    model.add(Conv2D(3, 3, padding='same'))
    model.add(Activation('sigmoid'))

    model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
    return model

simple_decoder = Simple_decoder()
simple_decoder.summary()
```

```python
# Let's now attempt to train the decoder with a single fixed value
history = simple_decoder.fit(np.array([1]), images[2:3], epochs=100, verbose=0)

plt.title('loss')
plt.plot(history.history['loss'])
plt.show()

plt.title('accuracy')
plt.plot(history.history['acc'])
plt.show()
```

```python
pred = simple_decoder.predict(np.array([1]))[0]

plt.imshow(pred)
plt.show()
```

```python
# yay! After 300 epoch we achieved a model that can pretty accurately decode an image stored in a single value!
```

```python
# let's now try, for the sake of the experiment, to train the network to decode any random value from a normal distribution
simple_decoder2 = Simple_decoder()

randoms = np.random.randn(100, 1)
repeated_image = np.repeat(images[2:3], 100, axis=0)

history = simple_decoder2.fit(randoms, repeated_image, epochs=100, verbose=0)

plt.title('loss')
plt.plot(history.history['loss'])
plt.show()

plt.title('accuracy')
plt.plot(history.history['acc'])
plt.show()

pred = simple_decoder2.predict(np.array([1]))[0]

plt.imshow(pred)
plt.show()
```

```python
pred = simple_decoder2.predict(np.random.randn(3, 1))

plt.imshow(pred[0])
plt.show()

plt.imshow(pred[1])
plt.show()

plt.imshow(pred[2])
plt.show()
```

```python
# our network seems to have learned to draw a cat from any random value in a normal distribution, yeah!
# let's now train the network to draw 2 different images: cat and dog, from fixed values...

simple_decoder3 = Simple_decoder()

history = simple_decoder3.fit(np.array([1, 2]), np.array([images[2], images[4]]), epochs=300, verbose=0)

plt.title('loss')
plt.plot(history.history['loss'])
plt.show()

plt.title('accuracy')
plt.plot(history.history['acc'])
plt.show()

pred = simple_decoder3.predict(np.array([1, 2]))

plt.imshow(pred[0])
plt.show()

plt.imshow(pred[1])
plt.show()
```

```python
# the network seems to have done a pretty good job reconstructing this 2 images, that's definitely awesome.
# but let's try to see what happens if I ask the network to reconstruct the image based in a number 
# in between cat and dog, and in some extreme numbers far from both

pred = simple_decoder3.predict(np.array([1.5, -1, 4]))

plt.imshow(pred[0])
plt.show()

plt.imshow(pred[1])
plt.show()

plt.imshow(pred[2])
plt.show()
```

```python
# let's now try to make our network lear from two normal distributions 
# with different mean and a little bit of overlapping
randoms2_1 = np.random.randn(50, 1) / 2 + 1 # mean in 1
randoms2_2 = np.random.randn(50, 1) / 2 + 2 # mean in 2
randoms2 = np.concatenate([randoms2_1, randoms2_2])

repeated_image2 = np.concatenate([np.repeat(images[2:3], 50, axis=0), np.repeat(images[4:5], 50, axis=0)])

simple_decoder4 = Simple_decoder()

history = simple_decoder4.fit(randoms2, repeated_image2, epochs=100, verbose=0, shuffle=True)

plt.title('loss')
plt.plot(history.history['loss'])
plt.show()

plt.title('accuracy')
plt.plot(history.history['acc'])
plt.show()

pred = simple_decoder4.predict(np.array([1, 2]))

plt.imshow(pred[0])
plt.show()

plt.imshow(pred[1])
plt.show()
```

```python
# It has a harder time reconstructing well defined images, let's now try with the values from before:

pred = simple_decoder4.predict(np.array([1.5, -1, 4]))

plt.imshow(pred[0])
plt.show()

plt.imshow(pred[1])
plt.show()

plt.imshow(pred[2])
plt.show()
```

```python
# hmm, not much of a difference...
# let's try now different architectures to make it converge faster...
```

```python
def Naive_decoder():
    model = Sequential()
    model.add(Dense(128*128*3, input_shape=(1,)))
    model.add(Activation('relu'))
    model.add(Reshape((128, 128, 3)))
    model.add(Conv2D(3, 3, padding='same'))
    model.add(Activation('sigmoid'))

    model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
naive_decoder = Naive_decoder()
naive_decoder.summary()

history = naive_decoder.fit(np.array([1, 2]), np.array([images[2], images[4]]), epochs=300, verbose=0)

plt.title('loss')
plt.plot(history.history['loss'])
plt.show()

plt.title('accuracy')
plt.plot(history.history['acc'])
plt.show()

pred = naive_decoder.predict(np.array([1, 2]))

plt.imshow(pred[0])
plt.show()

plt.imshow(pred[1])
plt.show()
```

```python
# Using a model too small gives bad results...
```

```python
def DeepCnn_decoder():
    model = Sequential()
    model.add(Dense(128*128*512, input_shape=(1,)))
    model.add(Activation('relu'))
    model.add(Reshape((128, 128, 512)))
    model.add(Conv2D(256, 3, padding='same'))
    model.add(Activation('relu'))
    model.add(Conv2D(128, 3, padding='same'))
    model.add(Activation('relu'))
    model.add(Conv2D(64, 3, padding='same'))
    model.add(Activation('relu'))
    model.add(Conv2D(32, 3, padding='same'))
    model.add(Activation('relu'))
    model.add(Conv2D(3, 3, padding='same'))
    model.add(Activation('sigmoid'))

    model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
deep_cnn_decoder = DeepCnn_decoder()
deep_cnn_decoder.summary()

history = deep_cnn_decoder.fit(np.array([1, 2]), np.array([images[2], images[4]]), epochs=50, verbose=0)

plt.title('loss')
plt.plot(history.history['loss'])
plt.show()

plt.title('accuracy')
plt.plot(history.history['acc'])
plt.show()

pred = deep_cnn_decoder.predict(np.array([1, 2]))

plt.imshow(pred[0])
plt.show()

plt.imshow(pred[1])
plt.show()
```

```python
def MicroCnn_decoder():
    model = Sequential()
    model.add(Dense(16*16*512, input_shape=(1,)))
    model.add(Activation('relu'))
    model.add(Reshape((16, 16, 512)))
    model.add(Conv2D(256, 3, padding='same'))
    model.add(Activation('relu'))
    model.add(Conv2D(128, 3, padding='same'))
    model.add(Activation('relu'))
    model.add(Conv2D(64, 3, padding='same'))
    model.add(Activation('relu'))
    model.add(UpSampling2D(4))
    model.add(Conv2D(32, 3, padding='same'))
    model.add(Activation('relu'))
    model.add(UpSampling2D(2))
    model.add(Conv2D(3, 3, padding='same'))
    model.add(Activation('sigmoid'))

    model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
micro_cnn_decoder = MicroCnn_decoder()
micro_cnn_decoder.summary()

history = micro_cnn_decoder.fit(np.array([1, 2]), np.array([images[2], images[4]]), epochs=150, verbose=0)

plt.title('loss')
plt.plot(history.history['loss'])
plt.show()

plt.title('accuracy')
plt.plot(history.history['acc'])
plt.show()

pred = micro_cnn_decoder.predict(np.array([1, 2]))

plt.imshow(pred[0])
plt.show()

plt.imshow(pred[1])
plt.show()
```
